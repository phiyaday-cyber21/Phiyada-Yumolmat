## 1. ทำไมต้องมี Function?

ลองดูปัญหานี้: ต้องการคำนวณ BMI ของนักศึกษา 3 คน

```python
# คนที่ 1
weight1, height1 = 65, 1.70
bmi1 = weight1 / (height1 ** 2)
print(f"BMI: {bmi1:.2f}")

# คนที่ 2
weight2, height2 = 58, 1.60
bmi2 = weight2 / (height2 ** 2)
print(f"BMI: {bmi2:.2f}")

# คนที่ 3
weight3, height3 = 80, 1.75
bmi3 = weight3 / (height3 ** 2)
print(f"BMI: {bmi3:.2f}")
```

สังเกตว่าโค้ด 3 ชุดนี้ **ทำสิ่งเดียวกัน** ซ้ำๆ แค่ค่า input ต่างกัน — ถ้ามีนักศึกษา 100 คน จะเขียนโค้ดซ้ำ 100 ชุดไหม?

- จัดการโค้ดให้เป็นระเบียบ แยกเป็นส่วนๆ ตามหน้าที่
- แก้ไขที่เดียว ใช้ได้ทุกที่ที่เรียก function นั้น
- ทดสอบและ debug ง่ายขึ้น (ทดสอบทีละ function)

## 2. การสร้างฟังก์ชันด้วย `def`

### โครงสร้างพื้นฐาน

```python
def ชื่อฟังก์ชัน():
    # โค้ดที่ต้องการทำงาน
```

**กฎสำคัญ:** เหมือน `if`/`for`/`while` — ต้องมี `:` ท้ายบรรทัด `def` และโค้ดด้านในต้องเยื้อง (indent)


````python
# ฟังก์ชันแบบง่ายที่สุด - ไม่มี parameter ไม่มี return
def greet():
    print("สวัสดี ยินดีต้อนรับสู่วิชา Essential Data Science")

# การ "สร้าง" ฟังก์ชันไม่ได้ทำให้โค้ดข้างในรัน ต้อง "เรียกใช้" ก่อน
greet()    # เรียกใช้ฟังก์ชัน (call function)
greet()    # เรียกใช้ซ้ำได้เรื่อยๆ
````

**ข้อสังเกตสำคัญ:** ตอนนิยามฟังก์ชัน (เขียน `def`) โค้ดข้างในจะ**ไม่รัน**ทันที — ต้อง "เรียกใช้" (call) ด้วยชื่อฟังก์ชันตามด้วย `()` ก่อน โค้ดข้างในจึงจะทำงาน


````python
def calculate_bmi_demo():
    weight = 65
    height = 1.70
    bmi = weight / (height ** 2)
    print(f"BMI ตัวอย่าง: {bmi:.2f}")

print("ก่อนเรียกใช้ฟังก์ชัน")
calculate_bmi_demo()   # ตรงนี้โค้ดข้างในฟังก์ชันถึงจะรัน
print("หลังเรียกใช้ฟังก์ชัน")
````

## 3. Parameter (พารามิเตอร์)

ฟังก์ชันด้านบนคำนวณ BMI ได้แค่ค่าเดียวที่ hardcode ไว้ — ถ้าอยากให้รับค่า weight/height ที่ต่างกันได้ ต้องใช้ **parameter**

**Parameter** คือ "ตัวแปรรับค่า" ที่ใส่ไว้ในวงเล็บตอนนิยามฟังก์ชัน ส่วน **Argument** คือ "ค่าจริง" ที่ส่งเข้าไปตอนเรียกใช้ฟังก์ชัน (คนละคำ แต่มักใช้ปนกันในบทสนทนาทั่วไป)

```python
def ชื่อฟังก์ชัน(parameter1, parameter2):
    # ใช้งาน parameter1, parameter2 ได้เหมือนตัวแปรปกติ
```


````python
def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)
    print(f"น้ำหนัก {weight} กก. ส่วนสูง {height} ม. -> BMI = {bmi:.2f}")

# เรียกใช้พร้อมส่ง argument เข้าไป
calculate_bmi(65, 1.70)    # weight=65, height=1.70
calculate_bmi(58, 1.60)    # ค่าต่างกัน แต่ใช้ฟังก์ชันเดียวกัน
calculate_bmi(80, 1.75)
````

### 3.1 การส่ง Argument แบบระบุชื่อ (Keyword Argument)

ปกติ argument จะถูกจับคู่กับ parameter ตามลำดับ (เรียกว่า positional argument) แต่ถ้าระบุชื่อ parameter ตรงๆ จะสลับลำดับได้ และโค้ดอ่านง่ายขึ้น


````python
def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)
    print(f"น้ำหนัก {weight} กก. ส่วนสูง {height} ม. -> BMI = {bmi:.2f}")

# Positional argument - ต้องเรียงลำดับให้ตรง
calculate_bmi(65, 1.70)

# Keyword argument - ระบุชื่อ parameter ตรงๆ สลับลำดับได้
calculate_bmi(height=1.70, weight=65)   # สลับลำดับได้เพราะระบุชื่อ
````

### 3.2 Parameter หลายตัว และเชื่อมกับ Collection ที่เรียนมาแล้ว

Parameter รับ collection (list, dict, tuple, set) เข้าไปได้เหมือนตัวแปรปกติทุกอย่าง — ใช้บ่อยมากในงานจริง


````python
def summarize_scores(scores):   # scores เป็น list ของคะแนน
    total = sum(scores)
    average = total / len(scores)
    print(f"คะแนนทั้งหมด: {scores}")
    print(f"ผลรวม: {total}, ค่าเฉลี่ย: {average:.2f}")

scores_list = [85, 90, 78, 92, 60]
summarize_scores(scores_list)
````

````python
def print_student_info(student_dict):   # รับ dict เข้ามาเป็น parameter
    print(f"ชื่อ: {student_dict['name']}")
    print(f"GPA: {student_dict['gpa']}")

student = {"name": "มานี", "gpa": 3.75}
print_student_info(student)
````

## 4. `return` — การคืนค่าจากฟังก์ชัน

ฟังก์ชันที่ผ่านมาแค่ `print()` ผลลัพธ์ออกมา แต่**ไม่ได้ "คืนค่า"** กลับไปให้โค้ดส่วนอื่นใช้ต่อ — ถ้าต้องการเอาผลลัพธ์ไปใช้ต่อ (เก็บในตัวแปร, เอาไปคำนวณต่อ) ต้องใช้ `return`

### ความแตกต่างระหว่าง `print()` กับ `return`

| | `print()` | `return` |
|---|---|---|
| ทำอะไร | แสดงผลลัพธ์บนหน้าจอ | ส่งค่ากลับไปให้ส่วนที่เรียกฟังก์ชัน |
| เอาไปใช้ต่อได้ไหม | ❌ ไม่ได้ (แค่แสดงผล) | ✅ ได้ (เก็บในตัวแปร, คำนวณต่อ) |
| ฟังก์ชันหยุดทำงานไหม | ❌ ไม่หยุด (โค้ดถัดไปยังรันต่อ) | ✅ หยุดทันที (ออกจากฟังก์ชันเลย) |


````python
def calculate_bmi_v1(weight, height):
    bmi = weight / (height ** 2)
    print(f"BMI = {bmi:.2f}")   # แค่ print ไม่ return

def calculate_bmi_v2(weight, height):
    bmi = weight / (height ** 2)
    return bmi   # คืนค่ากลับไป เอาไปใช้ต่อได้

# v1: เรียกแล้วเห็นผลลัพธ์ แต่เอาไปใช้ต่อไม่ได้
result1 = calculate_bmi_v1(65, 1.70)
print("result1 คือ:", result1)   # None! เพราะ v1 ไม่ได้ return อะไร

print()

# v2: เรียกแล้วได้ค่ากลับมา เอาไปใช้ต่อได้
result2 = calculate_bmi_v2(65, 1.70)
print("result2 คือ:", result2)         # ได้ค่า BMI จริง
print("เอาไปคำนวณต่อ:", result2 * 2)   # ใช้ต่อได้เลย
````

````python
x = calculate_bmi_v1(65, 1.70)
print(x)
````

````python
x = calculate_bmi_v2(65, 1.70)
print(x)
````

### 4.1 `return` หยุดการทำงานของฟังก์ชันทันที

โค้ดที่อยู่หลัง `return` ในฟังก์ชันจะไม่ถูกรันเลย — มีประโยชน์มากเวลาอยากให้ฟังก์ชัน "ออกก่อน" ถ้าเจอเงื่อนไขบางอย่าง


````python
def check_passing(score):
    if score < 0 or score > 100:
        return "คะแนนไม่ถูกต้อง"   # ออกจากฟังก์ชันทันทีถ้าคะแนนผิดปกติ
    if score >= 50:
        return "ผ่าน"
    return "ไม่ผ่าน"
    print("บรรทัดนี้ไม่มีทางรันถึง")   # unreachable code - ไม่มีทางถูกรัน

print(check_passing(75))
print(check_passing(30))
print(check_passing(150))
````

### 4.2 `return` หลายค่าพร้อมกัน (เชื่อมกับ tuple ที่เรียนมาแล้ว)

ฟังก์ชัน Python `return` ได้ค่าเดียว แต่ใส่หลายค่าคั่นด้วย `,` ได้ — จริงๆ แล้ว Python จะรวมเป็น **tuple** ให้อัตโนมัติ (ตรงกับเรื่อง tuple unpacking ที่เรียนใน Collections)


````python
def get_min_max_avg(numbers):
    return min(numbers), max(numbers), sum(numbers) / len(numbers)
    # จริงๆ คือ return (min(numbers), max(numbers), sum(numbers)/len(numbers))

scores = [85, 90, 78, 92, 60]

result = get_min_max_avg(scores)
print(result)              # ได้ tuple (60, 92, 81.0)
print(type(result))        # <class 'tuple'>

# unpack tuple ที่ return มา (เหมือนที่เรียนใน Notebook ก่อนหน้า)
lowest, highest, average = get_min_max_avg(scores)
print(f"ต่ำสุด: {lowest}, สูงสุด: {highest}, เฉลี่ย: {average:.2f}")
````

## 5. Default Argument (ค่าเริ่มต้นของพารามิเตอร์)

บางครั้งพารามิเตอร์ส่วนใหญ่มักมีค่าเดิมๆ ที่ใช้บ่อย — กำหนด **ค่าเริ่มต้น (default value)** ไว้ได้ ถ้าผู้เรียกฟังก์ชันไม่ส่งค่านั้นมา จะใช้ค่า default แทน

### โครงสร้าง

```python
def ชื่อฟังก์ชัน(parameter1, parameter2=ค่าเริ่มต้น):
    # ...
```

**กฎสำคัญ:** parameter ที่มี default value ต้องอยู่**หลัง**parameter ที่ไม่มี default เสมอ (เขียนสลับกันจะ error)


````python
def greet(name, greeting="สวัสดี"):   # greeting มีค่า default
    print(f"{greeting} {name}")

greet("มานี")                      # ไม่ส่ง greeting -> ใช้ค่า default "สวัสดี"
greet("สมชาย", "หวัดดี")           # ส่ง greeting มาเอง -> ใช้ค่าที่ส่งมาแทน
greet("วิชัย", greeting="Hello")   # ส่งแบบ keyword argument ก็ได้
````

### 5.1 ตัวอย่าง: ฟังก์ชันคำนวณเกรดที่ปรับเกณฑ์ได้ (เชื่อมกับ if/elif/else)

ใช้ default argument เพื่อให้ฟังก์ชันยืดหยุ่น — เปลี่ยนเกณฑ์ผ่านได้โดยไม่ต้องแก้โค้ดในฟังก์ชัน


````python
def get_grade(score, pass_threshold=50):
    if score >= 80:
        return "A"
    elif score >= 70:
        return "B"
    elif score >= 60:
        return "C"
    elif score >= pass_threshold:
        return "D"
    else:
        return "F"

print(get_grade(55))                       # ใช้เกณฑ์ผ่าน default = 50 -> ได้ D
print(get_grade(55, pass_threshold=60))    # เปลี่ยนเกณฑ์ผ่านเป็น 60 -> ได้ F แทน
````

### ⚠️ ข้อผิดพลาดที่พบบ่อย: ใช้ mutable object (list/dict) เป็นค่า default

**ห้ามใช้ list หรือ dict เป็นค่า default ตรง ๆ** เพราะ Python จะสร้าง object นั้นแค่ครั้งเดียวตอนนิยามฟังก์ชัน แล้วใช้ตัวเดิมซ้ำทุกครั้งที่เรียก — ทำให้เกิดบั๊กที่ไม่คาดคิด


````python
# ❌ วิธีที่ผิด - อย่าทำตามนี้!
def add_item_wrong(item, my_list=[]):   # list default ใช้ตัวเดิมซ้ำทุกครั้ง!
    my_list.append(item)
    return my_list

print(add_item_wrong("ปากกา"))    # ['ปากกา']
print(add_item_wrong("สมุด"))     # ['ปากกา', 'สมุด'] -- เห็นไหม! ค่าจากครั้งก่อนยังอยู่ ทั้งที่ไม่ได้ตั้งใจ
````

````python
# ✅ วิธีที่ถูก - ใช้ None เป็น default แทน แล้วสร้าง list ใหม่ข้างในฟังก์ชัน
def add_item_correct(item, my_list=None):
    if my_list is None:
        my_list = []   # สร้าง list ใหม่ทุกครั้งที่ไม่ได้ส่ง my_list มา
    my_list.append(item)
    return my_list

print(add_item_correct("ปากกา"))    # ['ปากกา']
print(add_item_correct("สมุด"))     # ['สมุด'] -- ถูกต้องแล้ว ไม่เก็บค่าจากครั้งก่อน
print(add_item_correct("ยางลบ"))
````

## 6. Scope (ขอบเขตของตัวแปร)

**Scope** คือ "พื้นที่" ที่ตัวแปรนั้นมองเห็นได้และใช้งานได้ — ตัวแปรที่สร้างในฟังก์ชันหนึ่ง จะ**มองไม่เห็นจากนอกฟังก์ชัน** (และในทางกลับกันก็มีข้อยกเว้น) เรื่องนี้สำคัญมากเพราะเป็นจุดที่มือใหม่สับสนบ่อย

### 6.1 Local Scope — ตัวแปรที่สร้างในฟังก์ชัน


````python
def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)   # bmi เป็น local variable - มีอยู่แค่ในฟังก์ชันนี้
    return bmi

result = calculate_bmi(65, 1.70)
print(result)

# ลองเข้าถึง bmi จากนอกฟังก์ชัน -> จะ error เพราะ bmi เป็น local variable
# print(bmi)   # ลบ # แล้วรันดู จะเห็น NameError
````

### 6.2 Global Scope — ตัวแปรที่สร้างนอกฟังก์ชัน (ระดับบนสุดของไฟล์)

ตัวแปรที่สร้างนอกฟังก์ชันทั้งหมด เรียกว่า **global variable** — ฟังก์ชันสามารถ "อ่าน" ค่าตัวแปร global ได้ แต่**แก้ไขไม่ได้โดยตรง** (ถ้าไม่ใช้ `global` keyword)


````python
course_name = "Essential Data Science"   # global variable

def show_course():
    print(f"กำลังเรียนวิชา: {course_name}")   # อ่านค่า global variable ได้

show_course()
print(course_name)   # เข้าถึงได้จากนอกฟังก์ชันด้วย เพราะเป็น global
````

### 6.3 จุดที่สับสนบ่อยที่สุด: การ "แก้ไข" ตัวแปร global จากในฟังก์ชัน

ถ้าในฟังก์ชันมีการ**กำหนดค่า** (`=`) ให้ตัวแปรชื่อเดียวกับ global variable — Python จะมองว่านี่คือ**ตัวแปร local ตัวใหม่** ไม่ใช่การแก้ไขตัว global!


````python
counter = 0   # global variable

def increment_wrong():
    counter = counter + 1   # ERROR! Python คิดว่า counter ทางขวาเป็น local (ยังไม่ถูกสร้าง)
    print(counter)

# increment_wrong()   # ลบ # แล้วรันดู จะเห็น UnboundLocalError
````

````python
counter = 0   # global variable

def increment_correct():
    global counter           # ประกาศว่า "จะใช้ตัวแปร counter ตัวเดียวกับข้างนอก"
    counter = counter + 1     # ตอนนี้แก้ไขตัว global ได้แล้ว
    print(counter)

increment_correct()   # 1
increment_correct()   # 2
increment_correct()   # 3
print("ค่าสุดท้ายของ counter:", counter)
````

**⚠️ ข้อแนะนำ:** การใช้ `global` ทำให้โค้ดเข้าใจยากและ debug ยากขึ้น (เพราะฟังก์ชันไปแก้ไขสิ่งที่อยู่นอกตัวมันเอง) **วิธีที่ดีกว่า** คือใช้ `return` ค่าใหม่ออกมา แล้วอัปเดตตัวแปร global ด้วยตัวเองนอกฟังก์ชัน


````python
counter = 0

def increment(current_value):     # รับค่าปัจจุบันเข้ามาเป็น parameter
    return current_value + 1       # คืนค่าใหม่ออกไป ไม่ไปแก้ global ตรงๆ

counter = increment(counter)   # อัปเดตตัวแปรนอกฟังก์ชันด้วยตัวเอง ชัดเจนกว่า
print(counter)   # 1
counter = increment(counter)
print(counter)   # 2
````

### 6.4 สรุป Scope

| Scope | สร้างที่ไหน | มองเห็นจากไหนได้บ้าง |
|---|---|---|
| **Local** | ในฟังก์ชัน (parameter หรือตัวแปรที่สร้างข้างใน) | เห็นได้แค่ในฟังก์ชันนั้นเท่านั้น |
| **Global** | นอกฟังก์ชันทั้งหมด (ระดับบนสุดของไฟล์) | ฟังก์ชันทุกตัว **อ่าน**ได้ แต่ต้องใช้ `global` keyword ถึงจะ**แก้ไข**ได้ |

> 💡 **แนวทางที่ดี:** พยายามให้ฟังก์ชันรับค่าผ่าน parameter และคืนค่าผ่าน return เสมอ หลีกเลี่ยงการพึ่งพา/แก้ไข global variable จากในฟังก์ชัน — จะทำให้โค้ดอ่านง่าย ทดสอบง่าย และไม่มีบั๊กที่ตามรอยยาก


````python
b = 4
def updete():
  b = 5
print("b = ", b)

c = 4
def updete2():
  global c
  c = 5
updete2()
print("c =", c)
````

### ข้อ 1: Function — BMI Calculator
เขียนฟังก์ชัน `calculate_bmi(weight, height)` ที่:
1. คำนวณ BMI และ **return** ค่า BMI (ไม่ใช่แค่ print)
2. มี default argument `height=1.65` (เผื่อไม่ระบุส่วนสูง)
3. ทดสอบเรียกใช้ทั้งแบบระบุครบและแบบไม่ระบุ height


````python
# เขียนคำตอบข้อ 1 ที่นี่
# 1.คำนวณ BMI และ return ค่า BMI

def calculate_bmi(weight, height = 1.65):
    bmi = weight / (height** 2)
    return bmi

result1 = calculate_bmi(45, 1.65)
print("ทดสอบเรียกใช้แบบระบุครบ:", result1)

result2 = calculate_bmi(45)
print("การทดสอบแบบไม่ระบุ height:", result2)
````

### ข้อ 2: Function — ทำงานกับ list/dict
เขียนฟังก์ชัน `find_top_scorer(scores_dict)` ที่รับ dict ในรูปแบบ `{"ชื่อ": คะแนน, ...}` แล้ว **return** ชื่อและคะแนนของคนที่ได้คะแนนสูงสุด (เป็น tuple)

ทดสอบกับ:
```python
scores = {"มานี": 85, "สมชาย": 90, "วิชัย": 78}
```


````python
# เขียนคำตอบข้อ 2 ที่นี่
scores = {"มานี": 85, "สมชาย": 90, "วิชัย": 78}

def find_top_scorer(scores_dict):
  max_score = -100
  max_key = None

  for k, v in scores_dict.items():
      if v > max_score:
         max_key = k
         max_score = v
  return (max_key, max_score)

key, score = find_top_scorer(scores)
print(key, "ได้คะแนน", score)
````

###ข้อ 3: Average

ให้สร้างฟังก์ชัน calculate_avg โดยจะรับลิสต์ของตัวเลข แล้วให้ผลลัพธ์ของ ค่าเฉลี่ย

````python
#เขียนคำตอบข้อที่ 3 ที่นี่
#def calculate_avg(num_list):
#
#  retur avg

def calculate_avg(num_list):
    total = 0
    for num in num_list:
      total += num
    avg = total / len(num_list)
    return avg

print("ค่าเฉลี่ยคือ", calculate_avg([55, 65, 75, 80, 70]))


````

###ข้อ 4: Min

ให้สร้างฟังก์ชัน find_min โดยจะรับลิสต์ของตัวเลข แล้วให้ผลลัพธ์ของ ค่าต่ำสุด

````python
#เขียนคำตอบข้อ 4 ที่นี่

def find_min(num_list):
  total = num_list[0]
  for num in num_list:
     if num < total:
          total = num
  return total

print("ค่าต่ำสุดคือ", find_min([55, 65, 75, 80, 70]))
````
